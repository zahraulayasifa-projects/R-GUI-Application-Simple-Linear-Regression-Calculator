**DATA DICTIONARY**

1. `hotel`: Jenis hotel ( "Resort Hotel" & "City Hotel").
2. `is_canceled`: Indikator apakah reservasi dibatalkan (0 untuk tidak dibatalkan, 1 untuk dibatalkan).
3. `lead_time`: Jumlah hari antara tanggal pemesanan dan tanggal kedatangan.
4. `arrival_date_year`: Tahun kedatangan tamu.
5. `arrival_date_month`: Bulan kedatangan tamu.
6. `arrival_date_week_number`: Minggu dalam tahun untuk kedatangan tamu.
7. `arrival_date_day_of_month`: Hari dalam bulan untuk kedatangan tamu.
8. `stays_in_weekend_nights`: Jumlah malam yang dihabiskan oleh tamu di akhir pekan.
9. `stays_in_week_nights`: Jumlah malam yang dihabiskan oleh tamu pada hari kerja.
10. `adults`: Jumlah tamu dewasa.
11. `children`: Jumlah anak-anak yang menemani tamu.
12. `babies`: Jumlah bayi yang menemani tamu.
13. `meal`: Jenis makanan yang dipesan oleh tamu.
14. `country`: Negara asal tamu.
15. `market_segment`: Segmen pasar di mana reservasi dibuat.
16. `distribution_channel`: Saluran distribusi yang digunakan untuk membuat reservasi.
17. `is_repeated_guest`: Indikator apakah tamu adalah tamu berulang (0 untuk bukan, 1 untuk ya).
18. `previous_cancellations`: Jumlah reservasi sebelumnya yang telah dibatalkan oleh tamu.
19. `previous_bookings_not_canceled`: Jumlah reservasi sebelumnya yang tidak dibatalkan oleh tamu.
20. `reserved_room_type`: Tipe kamar yang telah dipesan oleh tamu.
21. `assigned_room_type`: Tipe kamar yang sebenarnya diberikan kepada tamu.
22. `booking_changes`: Jumlah perubahan yang dilakukan pada reservasi.
23. `deposit_type`: Jenis deposit yang dibayarkan oleh tamu.
24. `agent`: ID agen perjalanan yang menangani reservasi.
25. `company`: ID perusahaan yang membuat reservasi.
26. `days_in_waiting_list`: Jumlah hari dalam daftar tunggu sebelum reservasi dikonfirmasi.
27. `customer_type`: Jenis tamu (misalnya, "Transient" atau "Contract").
28. `adr`: Tarif harian rata-rata (Average Daily Rate) untuk reservasi.
29. `required_car_parking_spaces`: Jumlah tempat parkir yang diminta oleh tamu.
30. `total_of_special_requests`: Jumlah permintaan khusus yang dibuat oleh tamu.
31. `reservation_status`: Status reservasi (misalnya, "Check-Out" atau "Canceled").
32. `reservation_status_date`: Tanggal terakhir status reservasi diperbarui.

```python
def percent_missing(dfc):
    percent_nan = 100* dfc.isnull().sum() / len(dfc)
    percent_nan = percent_nan[percent_nan>0].sort_values()
    return percent_nan
```

```python
mean = dfc['adr'].mean()
median = dfc['adr'].median()
q1 = dfc['adr'].quantile(0.25)
q3 = dfc['adr'].quantile(0.75)
print(f"Mean (Rata-rata): {mean}")
print(f"Median (Q2): {median}")
print(f"Quartile 1 (Q1): {q1}")
print(f"Quartile 3 (Q3): {q3}")
```

```python
def outliers(df, feature):
    q1 = dfc[feature].quantile(0.25)
    q3 = dfc[feature].quantile(0.75)
    iqr = q3 - q1
    upper_lmt = q3 + 1.5 * iqr
    lower_lmt = q1 - 1.5 * iqr
    return upper_lmt, lower_lmt

upper_lmt, lower_lmt = outliers(dfc, "adr")
print("Upper Limit: ", upper_lmt)
print("Lower Limit: ", lower_lmt)
```
